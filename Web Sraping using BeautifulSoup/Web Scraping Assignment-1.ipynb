```python
'''
5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link
'''
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.cnbc.com/world/?region=world"


response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("div", class_="Card-titleContainer")

headlines = []
times = []
news_links = []

for article in articles:
    headline = article.find("a", class_="Card-title").text.strip()
    
    time_tag = article.find("time")
    if time_tag:
        time = time_tag.text.strip()
    else:
        time = "N/A"
    
    news_link = "https://www.cnbc.com" + article.find("a", class_="Card-title")["href"]
    
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)

data = {
    "Headline": headlines,
    "Time": times,
    "News Link": news_links
}

df = pd.DataFrame(data)

print(df)

  
```

```python
'''
Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL
'''


def scrape_dineout_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        restaurants = []

        for restaurant in soup.find_all('div', class_='listing'):
            name = restaurant.find('div', class_='restnt-name').text.strip()
            cuisine = restaurant.find('span', class_='double-line-ellipsis').text.strip()
            location = restaurant.find('span', class_='restnt-loc').text.strip()
            ratings = restaurant.find('div', class_='rating-block').text.strip()
            image_url = restaurant.find('img', class_='thumb-img')['data-src']

            restaurants.append({
                'Restaurant Name': name,
                'Cuisine': cuisine,
                'Location': location,
                'Ratings': ratings,
                'Image URL': image_url
            })

        return restaurants
    else:
        print("Failed to fetch the page.")
        return None

if __name__ == "__main__":
    url = "https://www.dineout.co.in/delhi-restaurants"
    restaurant_data = scrape_dineout_data(url)

    if restaurant_data:
        df = pd.DataFrame(restaurant_data)
        print(df)
        df.to_csv('restaurant_data.csv', index=False)
        
```
